# BeetleCast 05 — Prediction versus actual damage

This notebook creates presentation-ready validation images.

For each labelled year it shows:

- **Actual image:** Sentinel-2 RGB from the label year, with supplied bark-beetle damage polygons outlined.
- **Model prediction:** risk scores for held-out stable-forest patches, using only information available before the label year.

Leakage-safe timing:

- 2022 outcome: predictors through 2021
- 2023 outcome: predictors through 2022

The label-year RGB is used **only for visual comparison**, never as a model input.

Outputs are saved to:

```text
outputs/visual_validation/
```


````python
from pathlib import Path
import re
import warnings

import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import xarray as xr
import rasterio
import rioxarray

from affine import Affine
from pyproj import CRS
from rasterio.features import rasterize
from rasterio.transform import from_origin

from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.model_selection import GroupShuffleSplit
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.metrics import average_precision_score, roc_auc_score

warnings.filterwarnings("ignore")

RANDOM_STATE = 42
PATCH_PIXELS = 5
TEST_FRACTION = 0.25
PLOT_PADDING_PATCHES = 8

cwd = Path.cwd()
if (cwd / "hackathon_data").exists():
    PROJECT_ROOT = cwd
elif (cwd.parent / "hackathon_data").exists():
    PROJECT_ROOT = cwd.parent
else:
    raise FileNotFoundError(
        "Could not find hackathon_data/. Launch Jupyter from project_root."
    )

RAW_ROOT = PROJECT_ROOT / "hackathon_data" / "raw"
PROCESSED_ROOT = PROJECT_ROOT / "hackathon_data" / "processed"
LULC_ROOT = PROJECT_ROOT / "outputs" / "lulc"
OUTPUT_ROOT = PROJECT_ROOT / "outputs" / "visual_validation"
OUTPUT_ROOT.mkdir(parents=True, exist_ok=True)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("OUTPUT_ROOT:", OUTPUT_ROOT)
````

## 1. Load the stable-forest modelling table

````python
table_path = LULC_ROOT / "beetlecast_training_table_with_lulc.csv"

assert table_path.exists(), (
    "Missing outputs/lulc/beetlecast_training_table_with_lulc.csv. "
    "Run notebook 04 first."
)

data = pd.read_csv(table_path)

baseline_features = [
    "ndvi_recent",
    "ndre_recent",
    "nbr_recent",
    "ndvi_prior",
    "ndre_prior",
    "nbr_prior",
    "ndvi_change",
    "ndre_change",
    "nbr_change",
    "ndvi_std",
    "ndre_std",
    "valid_fraction",
]

temporal_aef_features = [
    column for column in data.columns
    if column.startswith("aef_")
]

model_features = baseline_features + temporal_aef_features

required = {
    "label_year",
    "patch_row",
    "patch_col",
    "target",
    "spatial_block",
    "stable_forest",
} | set(model_features)

missing = required - set(data.columns)
assert not missing, f"Missing columns: {missing}"

# CSV may store booleans as strings.
if data["stable_forest"].dtype == object:
    data["stable_forest"] = (
        data["stable_forest"].astype(str).str.lower() == "true"
    )

stable = data[
    data["stable_forest"]
].dropna(
    subset=model_features
    + ["target", "spatial_block", "patch_row", "patch_col"]
).copy()

print("Stable-forest rows:", len(stable))
print("Model features:", len(model_features))
print(stable.groupby(["label_year", "target"]).size())
````

## 2. Reproduce the blocked split and train the selected model

Only the held-out test patches will be shown in the prediction maps.


````python
splitter = GroupShuffleSplit(
    n_splits=1,
    test_size=TEST_FRACTION,
    random_state=RANDOM_STATE,
)

train_idx, test_idx = next(
    splitter.split(
        stable,
        stable["target"],
        stable["spatial_block"],
    )
)

train = stable.iloc[train_idx].copy()
test = stable.iloc[test_idx].copy()

model = HistGradientBoostingClassifier(
    learning_rate=0.05,
    max_iter=350,
    max_leaf_nodes=31,
    min_samples_leaf=25,
    l2_regularization=2.0,
    random_state=RANDOM_STATE,
)

sample_weights = compute_sample_weight(
    class_weight="balanced",
    y=train["target"].astype(int),
)

model.fit(
    train[model_features],
    train["target"].astype(int),
    sample_weight=sample_weights,
)

test["predicted_risk"] = model.predict_proba(
    test[model_features]
)[:, 1]

print("Held-out rows:", len(test))
print(
    "Held-out PR-AUC:",
    round(
        average_precision_score(
            test["target"].astype(int),
            test["predicted_risk"],
        ),
        3,
    ),
)
print(
    "Held-out ROC-AUC:",
    round(
        roc_auc_score(
            test["target"].astype(int),
            test["predicted_risk"],
        ),
        3,
    ),
)
print(test.groupby(["label_year", "target"]).size())
````

## 3. Open the Sentinel-2 cube and reconstruct the grid

````python
zarr_paths = list(RAW_ROOT.rglob("cube.zarr"))
f3_candidates = [
    path for path in zarr_paths
    if "f3" in str(path).lower()
    or "bark" in str(path).lower()
    or "germany" in str(path).lower()
]

assert f3_candidates or zarr_paths, "No Sentinel-2 cube.zarr found"
CUBE_PATH = (f3_candidates or zarr_paths)[0]

ds = xr.open_zarr(CUBE_PATH, chunks="auto")

if float(ds.y.values[0]) < float(ds.y.values[-1]):
    ds = ds.sortby("y", ascending=False)

for band in ["B02", "B03", "B04"]:
    assert band in ds.data_vars, f"Missing RGB band {band}"

x = np.asarray(ds.x.values)
y = np.asarray(ds.y.values)

dx = float(abs(x[1] - x[0]))
dy = float(abs(y[1] - y[0]))

left = float(x.min() - dx / 2)
top = float(y.max() + dy / 2)

base_transform = from_origin(left, top, dx, dy)
patch_transform = base_transform * Affine.scale(
    PATCH_PIXELS,
    PATCH_PIXELS,
)

base_height = ds.sizes["y"]
base_width = ds.sizes["x"]
patch_height = base_height // PATCH_PIXELS
patch_width = base_width // PATCH_PIXELS

def infer_cube_crs(dataset):
    try:
        if dataset.rio.crs is not None:
            return CRS.from_user_input(dataset.rio.crs)
    except Exception:
        pass

    if "spatial_ref" in dataset:
        attrs = dataset["spatial_ref"].attrs
        for key in ("crs_wkt", "spatial_ref", "wkt"):
            value = attrs.get(key)
            if value:
                return CRS.from_user_input(value)

    raise ValueError("Could not infer Sentinel-2 CRS")

grid_crs = infer_cube_crs(ds)

print("Cube:", CUBE_PATH)
print("Patch grid:", patch_height, "x", patch_width)
print("CRS:", grid_crs)
````

## 4. Load the supplied bark-beetle polygons

````python
label_candidates = list(
    PROCESSED_ROOT.rglob("*labels*gpkg")
) + list(
    PROCESSED_ROOT.rglob("*labels*geojson")
)

if not label_candidates:
    label_candidates = list(
        PROJECT_ROOT.rglob("F3_bark_beetle_labels_in_aoi.gpkg")
    ) + list(
        PROJECT_ROOT.rglob("F3_bark_beetle_labels_in_aoi.geojson")
    )

assert label_candidates, "Could not find cleaned bark-beetle labels"

LABEL_PATH = label_candidates[0]
labels = gpd.read_file(LABEL_PATH)

if labels.crs is None:
    raise ValueError("Label file has no CRS")

labels = labels.to_crs(grid_crs)

if "label_year" not in labels.columns:
    source_columns = [
        column for column in labels.columns
        if "tile" in column.lower()
        or "name" in column.lower()
        or "file" in column.lower()
    ]
    assert source_columns, (
        "No label_year column and no filename-like column found."
    )

    source = labels[source_columns[0]].astype(str)
    labels["label_year"] = pd.to_numeric(
        source.str.extract(r"(2022|2023)")[0],
        errors="coerce",
    )

labels["label_year"] = labels["label_year"].astype("Int64")

print("Labels:", LABEL_PATH)
print(labels.groupby("label_year").size())
````

## 5. Build actual RGB images and actual damage masks

The RGB composite comes from the label year and is used only as the observed image displayed next to the prediction.


````python
def clean_reflectance(dataset, band):
    values = dataset[band].astype("float32")

    if "n_obs" in dataset:
        values = values.where(dataset["n_obs"] > 0)

    cutoff = np.datetime64("2022-01-25")
    values = xr.where(
        dataset.time >= cutoff,
        xr.where(values >= 1000, values - 1000, 0),
        values,
    )

    return values / 10000.0

def make_rgb_patch_composite(year):
    channels = []

    for band in ["B04", "B03", "B02"]:
        reflectance = clean_reflectance(ds, band)
        year_data = reflectance.sel(
            time=slice(f"{year}-01-01", f"{year}-12-31")
        )

        if year_data.sizes.get("time", 0) == 0:
            raise ValueError(f"No Sentinel-2 observations for {year}")

        median = year_data.median("time", skipna=True)

        patch = median.coarsen(
            y=PATCH_PIXELS,
            x=PATCH_PIXELS,
            boundary="trim",
        ).mean(skipna=True).compute().values

        channels.append(patch)

    return np.stack(channels, axis=-1)

def stretch_rgb(rgb, crop_mask=None):
    result = np.zeros_like(rgb, dtype="float32")

    for channel in range(3):
        values = rgb[..., channel]

        if crop_mask is not None:
            sample = values[crop_mask & np.isfinite(values)]
        else:
            sample = values[np.isfinite(values)]

        if sample.size == 0:
            continue

        low, high = np.nanpercentile(sample, [2, 98])
        if high <= low:
            high = low + 1e-6

        result[..., channel] = np.clip(
            (values - low) / (high - low),
            0,
            1,
        )

    return result

def rasterize_actual_damage(year):
    cohort = labels[
        labels["label_year"].astype("Int64") == int(year)
    ]

    shapes = [
        (geometry, 1)
        for geometry in cohort.geometry
        if geometry is not None and not geometry.is_empty
    ]

    return rasterize(
        shapes,
        out_shape=(patch_height, patch_width),
        transform=patch_transform,
        fill=0,
        default_value=1,
        dtype="uint8",
        all_touched=True,
    )

years = sorted(
    set(test["label_year"].dropna().astype(int))
    & set(labels["label_year"].dropna().astype(int))
)

rgb_by_year = {
    year: make_rgb_patch_composite(year)
    for year in years
}

actual_mask_by_year = {
    year: rasterize_actual_damage(year)
    for year in years
}

print("Prepared years:", years)
````

## 6. Convert held-out predictions to map grids

````python
risk_grid_by_year = {}
test_target_grid_by_year = {}

for year in years:
    risk_grid = np.full(
        (patch_height, patch_width),
        np.nan,
        dtype="float32",
    )
    target_grid = np.full(
        (patch_height, patch_width),
        np.nan,
        dtype="float32",
    )

    cohort_test = test[
        test["label_year"].astype(int) == int(year)
    ]

    rows = cohort_test["patch_row"].astype(int).to_numpy()
    cols = cohort_test["patch_col"].astype(int).to_numpy()

    risk_grid[rows, cols] = (
        cohort_test["predicted_risk"].to_numpy().astype("float32")
    )
    target_grid[rows, cols] = (
        cohort_test["target"].to_numpy().astype("float32")
    )

    risk_grid_by_year[year] = risk_grid
    test_target_grid_by_year[year] = target_grid

    print(
        year,
        "held-out patches:",
        int(np.isfinite(risk_grid).sum()),
    )
````

## 7. Create prediction-versus-actual images

Red outlines show the supplied mapped damage polygons.  
The risk panel contains only held-out test patches.


````python
def crop_bounds(mask, padding=PLOT_PADDING_PATCHES):
    rows, cols = np.where(mask)

    if len(rows) == 0:
        return 0, mask.shape[0], 0, mask.shape[1]

    r0 = max(0, int(rows.min()) - padding)
    r1 = min(mask.shape[0], int(rows.max()) + padding + 1)
    c0 = max(0, int(cols.min()) - padding)
    c1 = min(mask.shape[1], int(cols.max()) + padding + 1)

    return r0, r1, c0, c1

saved_figures = []

for year in years:
    risk_grid = risk_grid_by_year[year]
    actual_mask = actual_mask_by_year[year]

    display_region = (
        np.isfinite(risk_grid)
        | (actual_mask > 0)
    )

    r0, r1, c0, c1 = crop_bounds(display_region)

    rgb_crop = rgb_by_year[year][r0:r1, c0:c1]
    risk_crop = risk_grid[r0:r1, c0:c1]
    actual_crop = actual_mask[r0:r1, c0:c1]

    rgb_crop = stretch_rgb(
        rgb_crop,
        crop_mask=np.any(np.isfinite(rgb_crop), axis=-1),
    )

    cohort_test = test[
        test["label_year"].astype(int) == int(year)
    ].copy()

    threshold = cohort_test["predicted_risk"].quantile(0.90)

    fig, axes = plt.subplots(
        1,
        2,
        figsize=(15, 7),
        constrained_layout=True,
    )

    # Actual observed image.
    axes[0].imshow(rgb_crop)
    if actual_crop.max() > 0:
        axes[0].contour(
            actual_crop,
            levels=[0.5],
            colors="red",
            linewidths=1.4,
        )

    axes[0].set_title(
        f"Actual {year} satellite image\n"
        "Supplied mapped damage outlined in red",
        fontsize=13,
    )
    axes[0].axis("off")

    # Held-out risk.
    axes[1].imshow(rgb_crop, alpha=0.35)
    risk_image = axes[1].imshow(
        risk_crop,
        vmin=0,
        vmax=1,
        cmap="magma",
        alpha=np.where(np.isfinite(risk_crop), 0.92, 0),
    )

    if actual_crop.max() > 0:
        axes[1].contour(
            actual_crop,
            levels=[0.5],
            colors="cyan",
            linewidths=1.2,
        )

    high_risk = (
        np.isfinite(risk_crop)
        & (risk_crop >= threshold)
    )

    if high_risk.any():
        axes[1].contour(
            high_risk.astype(int),
            levels=[0.5],
            colors="white",
            linewidths=0.8,
        )

    axes[1].set_title(
        f"BeetleCast held-out prediction\n"
        f"Inputs through {year - 1}; top 10% outlined in white",
        fontsize=13,
    )
    axes[1].axis("off")

    colorbar = fig.colorbar(
        risk_image,
        ax=axes[1],
        fraction=0.046,
        pad=0.03,
    )
    colorbar.set_label("Predicted canopy-damage risk")

    red_patch = mpatches.Patch(
        color="red",
        label="Actual mapped damage",
    )
    cyan_patch = mpatches.Patch(
        color="cyan",
        label="Actual outline on prediction",
    )
    white_patch = mpatches.Patch(
        color="white",
        label="Highest-risk 10% of held-out patches",
    )

    fig.legend(
        handles=[red_patch, cyan_patch, white_patch],
        loc="lower center",
        ncol=3,
        frameon=True,
    )

    fig.suptitle(
        f"BeetleCast: predicted risk versus observed mapped damage — {year}",
        fontsize=16,
    )

    output_path = (
        OUTPUT_ROOT
        / f"beetlecast_{year}_prediction_vs_actual.png"
    )

    plt.savefig(
        output_path,
        dpi=240,
        bbox_inches="tight",
    )
    plt.show()

    saved_figures.append(output_path)
    print("Saved:", output_path)
````

## 8. Create a single combined demo image

This combines both labelled years into one figure for a slide or final presentation.


````python
fig, axes = plt.subplots(
    len(years),
    2,
    figsize=(15, 7 * len(years)),
    constrained_layout=True,
)

if len(years) == 1:
    axes = np.array([axes])

last_risk_image = None

for row_index, year in enumerate(years):
    risk_grid = risk_grid_by_year[year]
    actual_mask = actual_mask_by_year[year]

    display_region = (
        np.isfinite(risk_grid)
        | (actual_mask > 0)
    )

    r0, r1, c0, c1 = crop_bounds(display_region)

    rgb_crop = rgb_by_year[year][r0:r1, c0:c1]
    risk_crop = risk_grid[r0:r1, c0:c1]
    actual_crop = actual_mask[r0:r1, c0:c1]

    rgb_crop = stretch_rgb(
        rgb_crop,
        crop_mask=np.any(np.isfinite(rgb_crop), axis=-1),
    )

    cohort_test = test[
        test["label_year"].astype(int) == int(year)
    ]
    threshold = cohort_test["predicted_risk"].quantile(0.90)

    axes[row_index, 0].imshow(rgb_crop)

    if actual_crop.max() > 0:
        axes[row_index, 0].contour(
            actual_crop,
            levels=[0.5],
            colors="red",
            linewidths=1.2,
        )

    axes[row_index, 0].set_title(
        f"Actual {year} image and mapped damage",
        fontsize=13,
    )
    axes[row_index, 0].axis("off")

    axes[row_index, 1].imshow(rgb_crop, alpha=0.35)
    last_risk_image = axes[row_index, 1].imshow(
        risk_crop,
        vmin=0,
        vmax=1,
        cmap="magma",
        alpha=np.where(np.isfinite(risk_crop), 0.92, 0),
    )

    if actual_crop.max() > 0:
        axes[row_index, 1].contour(
            actual_crop,
            levels=[0.5],
            colors="cyan",
            linewidths=1.1,
        )

    high_risk = (
        np.isfinite(risk_crop)
        & (risk_crop >= threshold)
    )

    if high_risk.any():
        axes[row_index, 1].contour(
            high_risk.astype(int),
            levels=[0.5],
            colors="white",
            linewidths=0.8,
        )

    axes[row_index, 1].set_title(
        f"Held-out risk using data through {year - 1}",
        fontsize=13,
    )
    axes[row_index, 1].axis("off")

if last_risk_image is not None:
    colorbar = fig.colorbar(
        last_risk_image,
        ax=axes[:, 1],
        fraction=0.025,
        pad=0.02,
    )
    colorbar.set_label("Predicted canopy-damage risk")

fig.suptitle(
    "BeetleCast — leakage-safe prediction versus observed mapped damage",
    fontsize=17,
)

combined_path = (
    OUTPUT_ROOT
    / "beetlecast_prediction_vs_actual_combined.png"
)

plt.savefig(
    combined_path,
    dpi=240,
    bbox_inches="tight",
)
plt.show()

print("Saved:", combined_path)
````

## 9. Export the highest-risk held-out patches

````python
top_predictions = (
    test.sort_values(
        ["label_year", "predicted_risk"],
        ascending=[True, False],
    )
    .groupby("label_year")
    .head(20)
    [
        [
            "label_year",
            "patch_row",
            "patch_col",
            "target",
            "predicted_risk",
        ]
    ]
)

top_path = (
    OUTPUT_ROOT
    / "beetlecast_top_heldout_predictions.csv"
)
top_predictions.to_csv(top_path, index=False)

display(top_predictions)
print("Saved:", top_path)
print("\nVisual validation complete.")
````
